```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors

# Definir la función para calcular descriptores moleculares
def getMolDescriptors(mol, missingVal=None):
    '''Calcula la lista completa de descriptores para una molécula.
       missingVal se utiliza si el descriptor no puede ser calculado.'''
    res = {}
    for nm, fn in Descriptors.descList:
        try:
            val = fn(mol)
        except:
            val = missingVal
        res[nm] = val
    return res

# Ruta del archivo SMILES
ruta_smiles = "/home/angel/Escritorio/smiles.txt"  # Reemplaza "usuario" con tu nombre real

# Leer el archivo SMILES y extraer identificadores y estructuras
with open(ruta_smiles, "r") as file:
    smiles_data = [line.strip().split(" ", 1) for line in file.readlines()]

# Lista para almacenar los resultados
resultados = []

# Procesar cada SMILES y calcular descriptores
for i, (smiles, identificador) in enumerate(smiles_data):
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        descriptors = getMolDescriptors(mol)
        descriptors["Identificador"] = identificador  # Agregar el identificador de la molécula
        resultados.append(descriptors)
        print(f"Molécula {i + 1} ({identificador}): {descriptors}")
    else:
        print(f"Molécula {i + 1} ({identificador}) no se pudo procesar.")

# Crear un DataFrame y eliminar columnas con valores nulos o ceros
df = pd.DataFrame(resultados)
df = df.loc[:, (df != 0).any(axis=0)]  # Eliminar columnas con solo ceros
df = df.dropna(axis=1, how='all')  # Eliminar columnas completamente vacías

# Guardar en un archivo CSV
df.to_csv("descriptores_moleculares.csv", index=False)
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors

# Ruta del archivo SMILES
ruta_smiles = "/home/angel/Escritorio/smiles.txt"  

# Cargar el archivo de SMILES
try:
    with open(ruta_smiles, 'r') as f:
        # Leer las líneas y separar los SMILES de los identificadores
        smiles_data = [line.strip().split(" ", 1) for line in f if line.strip()]

    if not smiles_data:
        print("El archivo está vacío o no se ha leído correctamente.")
        exit()

except FileNotFoundError:
    print(f"No se encontró el archivo en la ruta: {ruta_smiles}")
    exit()

# Inicializar listas para guardar los datos
hc_ratios = []
oc_ratios = []
molecular_weights = []
identificadores = []

# Calcular H/C y O/C, además del peso molecular
for smiles, identificador in smiles_data:
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        mol_with_hydrogens = Chem.AddHs(mol)
        
        # Contar átomos de C, H y O
        num_c = sum(1 for atom in mol_with_hydrogens.GetAtoms() if atom.GetSymbol() == 'C')
        num_h = sum(1 for atom in mol_with_hydrogens.GetAtoms() if atom.GetSymbol() == 'H')
        num_o = sum(1 for atom in mol_with_hydrogens.GetAtoms() if atom.GetSymbol() == 'O')

        # Calcular relaciones H/C y O/C
        hc_ratio = num_h / num_c if num_c != 0 else None
        oc_ratio = num_o / num_c if num_c != 0 else None
        
        # Obtener el peso molecular
        mw = Descriptors.MolWt(mol_with_hydrogens)

        # Guardar los valores
        hc_ratios.append(hc_ratio)
        oc_ratios.append(oc_ratio)
        molecular_weights.append(mw)
        identificadores.append(identificador)  # Guardar el identificador de la molécula
    else:
        print(f"SMILES no válido: {smiles}")

# Crear un DataFrame sin la columna de SMILES
df = pd.DataFrame({
    'Identificadores': identificadores,
    'H/C': hc_ratios,
    'O/C': oc_ratios,
    'Molecular Weight': molecular_weights
})

# Eliminar columnas con solo valores nulos
df = df.dropna(axis=1, how='all')

# Guardar en un archivo CSV con el nombre solicitado
df.to_csv("relacion_HC.csv", index=False)

# Verificar si el DataFrame está vacío
if df.empty:
    print("El DataFrame está vacío, verifica los SMILES o el archivo de entrada.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import Descriptors
import matplotlib.cm as cm

# Cargar el archivo de SMILES
file_path = '/home/angel/Escritorio/smiles.txt'
try:
    with open(file_path, 'r') as f:
        # Leer las líneas y separar los SMILES de los identificadores
        smiles_list = [line.strip().split() for line in f if line.strip()]
except FileNotFoundError:
    print(f"El archivo '{file_path}' no se encontró. Verifica la ruta.")
    exit()

# Inicializar listas para guardar los datos
hc_ratios = []
oc_ratios = []
molecular_weights = []
identifiers = []

# Clasificar las moléculas según el identificador
for i, (smile, code) in enumerate(smiles_list, start=1):
    mol = Chem.MolFromSmiles(smile)
    if mol:
        mol_with_hydrogens = Chem.AddHs(mol)
        
        # Contar átomos de C, H y O
        num_c = sum(1 for atom in mol_with_hydrogens.GetAtoms() if atom.GetSymbol() == 'C')
        num_h = sum(1 for atom in mol_with_hydrogens.GetAtoms() if atom.GetSymbol() == 'H')
        num_o = sum(1 for atom in mol_with_hydrogens.GetAtoms() if atom.GetSymbol() == 'O')
        
        # Calcular relaciones H/C y O/C
        hc_ratio = num_h / num_c if num_c != 0 else 0
        oc_ratio = num_o / num_c if num_c != 0 else 0
        
        # Obtener el peso molecular
        mw = Descriptors.MolWt(mol_with_hydrogens)
        
        # Clasificar por familias químicas
        if 1 <= i <= 250:
            family = 'Flavonoids'
        elif 251 <= i <= 350:
            family = 'Xanthophylls'
        elif 351 <= i <= 400:
            family = 'Chlorophylls'
        else:
            family = 'Anthocyanins'
        
        # Guardar los valores
        hc_ratios.append(hc_ratio)
        oc_ratios.append(oc_ratio)
        molecular_weights.append(mw)
        identifiers.append(family)
    else:
        print(f"SMILES no válido: {smile}")

# Crear un DataFrame para los datos
df = pd.DataFrame({
    'Identifiers': identifiers,
    'H/C': hc_ratios,
    'O/C': oc_ratios,
    'Molecular Weight': molecular_weights
})

# Guardar los datos en un archivo CSV
csv_filename = '/home/angel/Escritorio/datos_graficos.csv'
df.to_csv(csv_filename, index=False)
print(f"Archivo guardado como {csv_filename}")

# Asignar colores únicos a cada familia química
families = df['Identifiers'].unique()
color_map = cm.get_cmap('viridis', len(families))
colors = {family: color_map(i / len(families)) for i, family in enumerate(families)}

# Gráfica 1: Van Krevelen (H/C vs O/C)
plt.figure(figsize=(10, 6))
for family in families:
    subset = df[df['Identifiers'] == family]
    plt.scatter(subset['O/C'], subset['H/C'], color=colors[family], label=family, edgecolors='k', alpha=0.7)

plt.title('Van Krevelen Diagram: O/C vs H/C')
plt.xlabel('O/C')
plt.ylabel('H/C')
plt.grid()
plt.axhline(0, color='grey', lw=1)
plt.axvline(0, color='grey', lw=1)
plt.legend(title='Chemical Family')
plt.show()

# Gráfica 2: H/C vs Peso Molecular
plt.figure(figsize=(10, 6))
for family in families:
    subset = df[df['Identifiers'] == family]
    plt.scatter(subset['Molecular Weight'], subset['H/C'], color=colors[family], label=family, edgecolors='k', alpha=0.7)

plt.title('H/C vs Molecular Weight')
plt.xlabel('Molecular Weight (g/mol)')
plt.ylabel('H/C')
plt.grid()
plt.axhline(0, color='grey', lw=1)
plt.axvline(0, color='grey', lw=1)
plt.legend(title='Chemical Family')
plt.show()
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs
import matplotlib.pyplot as plt
import numpy as np

# Ruta del archivo de SMILES
ruta_smiles = '/home/angel/Escritorio/smiles.txt'

# Cargar los SMILES y asignar familia química según la línea en la que aparecen
smiles_list = []
families = []
line_number = 0

with open(ruta_smiles, 'r') as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) == 2:
            smiles, code = parts
            smiles_list.append((smiles, code))
            
            # Determinar la familia basándose en la posición en la lista
            line_number += 1
            if 1 <= line_number <= 250:
                families.append('Flavonoides')
            elif 251 <= line_number <= 350:
                families.append('Xantofilas')
            elif 351 <= line_number <= 400:
                families.append('Clorofilas')
            elif 401 <= line_number <= 600:
                families.append('Antocianinas')

# Generación de fingerprints
fingerprints = []
valid_smiles = []
valid_families = []

for (smiles, family) in zip(smiles_list, families):
    mol = Chem.MolFromSmiles(smiles[0])  # Convertimos el SMILES en una molécula
    if mol:
        fp = AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048)
        fingerprints.append(fp)
        valid_smiles.append(smiles[0])
        valid_families.append(family)

# Cálculo de coeficientes de Tanimoto
tanimoto_scores = []
for i in range(len(fingerprints)):
    for j in range(i + 1, len(fingerprints)):  # Comparaciones únicas
        tanimoto = DataStructs.TanimotoSimilarity(fingerprints[i], fingerprints[j])
        tanimoto_scores.append(tanimoto)

# Graficar el histograma
n, bins, patches = plt.hist(tanimoto_scores, bins=50, color='lightgray', edgecolor='black')

# Resaltar barras según el coeficiente de Tanimoto
for i, patch in enumerate(patches):
    if bins[i] > 0.8:
        patch.set_facecolor('red')
    elif bins[i] > 0.6:
        patch.set_facecolor('orange')
    elif bins[i] > 0.4:
        patch.set_facecolor('yellow')

plt.xlabel('Tanimoto Coefficient (Tc)')
plt.ylabel('Frequency')
plt.title('Tanimoto coefficient distribution')

# Calcular F(Tc > s) para distintos umbrales
s_values = [0.3, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9]
F_values = []
total_pairs = len(tanimoto_scores)

for s in s_values:
    count = sum(1 for t in tanimoto_scores if t > s)
    F = count / total_pairs
    F_values.append(F)

# Crear una tabla con los valores calculados
prob_table = pd.DataFrame({
    'Threshold (s)': s_values,
    'F(Tc > s)': [f'{F:.1e}' for F in F_values]
})

# Agregar la tabla a la gráfica
table_text = 'Threshold (s)    F(Tc > s)\n' + '\n'.join(
    [f"{row['Threshold (s)']}          {row['F(Tc > s)']}" for _, row in prob_table.iterrows()]
)
plt.figtext(0.6, 0.4, table_text, fontsize=10, bbox=dict(facecolor='white', alpha=0.8))

plt.show()

# Guardar la matriz de similitud en un CSV
df_tanimoto = pd.DataFrame(tanimoto_scores, columns=['Tanimoto Coefficient'])
df_tanimoto.to_csv('/home/angel/Escritorio/datos_tanimoto.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs

# Configuración de Seaborn
sns.set(style="whitegrid")

# Ruta del archivo SMILES
archivo_smiles = "/home/angel/Escritorio/smiles.txt"

# Leer los primeros 250 SMILES (Flavonoides)
smiles_list = []
with open(archivo_smiles, "r") as file:
    for i, line in enumerate(file):
        if i >= 250:  # Solo los primeros 250
            break
        smiles, code = line.strip().split()
        smiles_list.append(smiles)

# Convertir SMILES a huellas moleculares
mols = [Chem.MolFromSmiles(s) for s in smiles_list if Chem.MolFromSmiles(s)]
fps = [AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048) for mol in mols]

# Crear matriz de similitud Tanimoto
n = len(fps)
tanimoto_matrix = np.zeros((n, n))

for i in range(n):
    for j in range(i, n):  # Evita cálculos duplicados
        sim = DataStructs.FingerprintSimilarity(fps[i], fps[j])
        tanimoto_matrix[i, j] = sim
        tanimoto_matrix[j, i] = sim  # Matriz simétrica

# Aplanar matriz y eliminar valores NaN o redundantes
tanimoto_values = tanimoto_matrix[np.triu_indices(n, k=1)]

# Graficar distribución de coeficientes de Tanimoto
plt.figure(figsize=(10, 6))
sns.histplot(tanimoto_values, bins=np.arange(0, 1.05, 0.05), 
             color='#66FF99', edgecolor='black', kde=True)

plt.title('Distribución de Coeficientes de Tanimoto - Flavonoides', fontsize=14, weight='bold')
plt.xlabel('Coeficiente de Tanimoto (Tc)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.xlim(0, 1)

# Cuadrícula y estilo
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Mostrar gráfico
plt.show()
```
